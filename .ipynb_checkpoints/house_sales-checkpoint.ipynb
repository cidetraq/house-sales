```python
import pandas as pd
from matplotlib import pyplot
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
%matplotlib inline

data=pd.read_csv('AmesHousing.tsv', sep='\t')
```

```python
def transform_features(df):
    train=df[0:1460]
    length=len(train)
    nulls=train.isnull().sum()
    cols=nulls[nulls>length*.25].index
    train=train.drop(labels=cols, axis=1)
    train=train.drop(labels=['Yr Sold'], axis=1)
    train['years_until_remod']=train['Year Remod/Add']-train['Year Built']
    
    nulls=train.isnull().sum()
    cols=nulls[(nulls<length*.05) & (nulls>0)].index
    num_df=train[cols].select_dtypes(include=['int', 'float', 'double'])
    train[num_df.columns]=num_df.fillna(num_df.mode())
    print(train[num_df.columns].isnull().sum())
    
    train['porch_area']=train['Screen Porch']+train['3Ssn Porch']+train['Enclosed Porch']+train['Open Porch SF']
    return train
```

```python
def select_features(df):
    return df[['Gr Liv Area','SalePrice']]
```

```python
def train_and_test(df):
    train=df[0:1460]
    test=df[1460:]
    lr=LinearRegression()
    cols=df[select_features.index].select_dtypes(include=['int', 'float', 'double'])
    lr.fit(df[cols], df['SalePrice'])
    predictions=lr.predict(test[cols])
    mse=mean_squared_error(test['SalePrice'], predictions)
    rmse=mse**(1/2)
    return rmse
    
```
